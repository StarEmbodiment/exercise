#### 4.  (1)建立一个全0矩阵 a, 大小为 3x3; 类型为整型（提示: dtype = int）(2)建立一个全1矩阵b,大小为4x5;  (3)建立一个单位矩阵c ,大小为4x4; (4)生成一个随机数矩阵d,大小为 3x2.

```python
matrix_a = np.zeros((3, 3), dtype=int)
print("全零矩阵:\n", matrix_a)

matrix_b = np.ones((4, 5))
print("全1矩阵:\n", matrix_b)

matrix_c = np.eye(4)
print("单位矩阵:\n", matrix_c)

matrix_d = np.random.random((3, 2))
print("随机矩阵:\n", matrix_d)
```

#### 24,画图，y=x*x 其中 x = np.arange(0, 100, 0.1) （提示这里用到  matplotlib.pyplot 库）

```python
import matplotlib.pyplot as plt
x_vals = np.arange(0, 100, 0.1)
y_vals = x_vals ** 2
plt.plot(x_vals, y_vals)
plt.title("y = x^2")
plt.show()
```

#### 25.画图。画正弦函数和余弦函数， x = np.arange(0, 3 * np.pi, 0.1)(提示：这里用到 np.sin() np.cos() 函数和 matplotlib.pyplot 库)

```python
x_vals = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x_vals)
y_cos = np.cos(x_vals)
plt.plot(x_vals, y_sin, label='sin')
plt.plot(x_vals, y_cos, label='cos')
plt.legend()
plt.title("Sine and Cosine")
plt.show()
```
